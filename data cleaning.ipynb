```python
import json


def clean_data(data):
    #removing blank users name
    data["users"]= [i for i in data["users"] if i["name"].strip()]  #[expression(x*X) for variable(x) in data ] 
    
    # removing duplicate friends name 
    for i in data["users"]: 
        i["friends"]=list(set(i["friends"]))   #here set has no duplicate value so we are converts a list havings friends into set and converting it back again into list
     
    # removing users(inactive) who either don't have friends or liking pages  
    data["users"]=[i for i in data["users"] if i["friends"] or i["liked_pages"]]   #removing users who does not have any frined or liking pages 

    #removing duplicate pages 
    unique={}
    for j in data["pages"]:
        unique[j["id"]]=j   #not  written as  j= unique[j["id"]] as then we will be accessing key before storing the value 
    data["pages"]=list(unique.values())    
        
    
    return data

data=json.load(open("data2.json"))
data=clean_data(data)
json.dump(data,open("delhi coders(2)","w"),indent=3)    ##[json.dump(obj,file name,operation,indent(spacing is provided)]
print("data is succesfully being cleaned ")
```
